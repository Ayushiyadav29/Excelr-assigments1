# Hclustering (Crime dataset)

```python
!pip install numpy==1.21.4
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
Crime=pd.read_csv('crime_data.csv',index_col=0)
```

```python
Crime
```

```python
#Normalise data
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(Crime.iloc[:,1:]))
data
```

```python
#Standarise data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Crime_subset=pd.DataFrame(scaler.fit_transform(Crime.iloc[:,1:]))
```

```python
Crime_subset
```

```python
#Distance,linkages
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(Crime_subset,method='average',metric='euclidean')
plt.figure(figsize=(10, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
#AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=4 ,linkage='average',affinity='euclidean').fit(Crime_subset)
```

```python
h_complete
```

```python
#Labels
Crime['clust']=h_complete.labels_
```

```python
Crime
```

```python
#groupdata by cluster values
Crime.iloc[:,1:].groupby(Crime.clust).mean()
```

```python
#check values according to clusters
data =Crime[(Crime.clust==1)]
data   
```

```python
data =Crime[(Crime.clust==2)]
data   
```

```python
data =Crime[(Crime.clust==0)]
data   
```


# Hclustering (Eastwestairlines dataset)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
Airlines=pd.read_excel('EastWestAirlines (1).xlsx',sheet_name='data')
Airlines
```

```python
#normalise data
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(Airlines.iloc[:,1:]))
data
```

```python
#Standarise data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Airlines_subset=pd.DataFrame(scaler.fit_transform(Airlines.iloc[:,1:]))
Airlines_subset
```

```python
#dendrogram
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(data,method='complete',metric='euclidean')
plt.figure(figsize=(10, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python

#AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=4 ,linkage='average',affinity='euclidean').fit(Airlines_subset)
```

```python
h_complete
```

```python
#labels
Airlines['clust']=h_complete.labels_
```

```python
Airlines
```

```python
#groupdata by cluster values
Airlines.iloc[:,1:].groupby(Airlines.clust).mean()
```

```python
#check values according to clusters
data =Airlines[(Airlines.clust==1)]
data   
```

```python
data =Airlines[(Airlines.clust==2)]
data
```
