# Naive bayes(Salary dataset)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
#train dataset
df_train=pd.read_csv('SalaryData_Test.csv.crdownload')
```

```python
df_train
```

```python
df1_train = df_train.drop_duplicates()
df1_train
import warnings
warnings.filterwarnings('ignore')
```

```python
df1_train.describe()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df1_train['workclass']= label_encoder.fit_transform(df1_train['workclass']) 
df1_train['education']= label_encoder.fit_transform(df1_train['education'])
df1_train['maritalstatus']= label_encoder.fit_transform(df1_train['maritalstatus'])
df1_train['occupation']= label_encoder.fit_transform(df1_train['occupation'])
df1_train['relationship']= label_encoder.fit_transform(df1_train['relationship'])
df1_train['race']= label_encoder.fit_transform(df1_train['race'])
df1_train['sex']= label_encoder.fit_transform(df1_train['sex'])
df1_train['native']= label_encoder.fit_transform(df1_train['native'])
df1_train['Salary']= label_encoder.fit_transform(df1_train['Salary'])
```

```python
df1_train.info()
```

```python
#testing datset
df_test = pd.read_csv('SalaryData_Test.csv.crdownload')
df_test
```

```python
df1_test = df_test.drop_duplicates()
df1_test
```

```python
#Label Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df1_test['workclass']= label_encoder.fit_transform(df1_test['workclass']) 
df1_test['education']= label_encoder.fit_transform(df1_test['education'])
df1_test['maritalstatus']= label_encoder.fit_transform(df1_test['maritalstatus'])
df1_test['occupation']= label_encoder.fit_transform(df1_test['occupation'])
df1_test['relationship']= label_encoder.fit_transform(df1_test['relationship'])
df1_test['race']= label_encoder.fit_transform(df1_test['race'])
df1_test['sex']= label_encoder.fit_transform(df1_test['sex'])
df1_test['native']= label_encoder.fit_transform(df1_test['native'])
df1_test['Salary']= label_encoder.fit_transform(df1_test['Salary'])
```

```python
df1_test.info()
```

```python
#Applying Train and Test split on Salary Dataset:
x_train = df1_train.iloc[:,0:13]
y_train = df1_train.iloc[:,-1] 
```

```python
x_test = df1_test.iloc[:,0:13]
y_test = df1_test.iloc[:,-1]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#Naive Bayes Classification :
#1. MultinomialNB
# Preparing a Multinomial naive bayes model on training data set 
from sklearn.naive_bayes import MultinomialNB as MB    # Multinomial Naive Bayes
```

```python
clf=MultinomialNB()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
#2.CategoricalNB
# Preparing a Categorical naive bayes model on training data set 
from sklearn.naive_bayes import CategoricalNB as CNB
classifier_cnb = CNB()
classifier_cnb.fit(x_train,y_train)
```

```python
# Model Accuracy on train set
train_pred_cnb = classifier_cnb.predict(x_train)
accuracy_train_cnb = np.mean(train_pred_cnb==y_train) 
```

```python
# Model Accuracy on test set
test_pred_cnb = classifier_cnb.predict(x_train)
accuracy_test_cnb = np.mean(test_pred_cnb==y_train)
```

```python
accuracy_train_cnb
```

```python
accuracy_test_cnb
```

```python
#3. GaussianNB
# Preparing a Gaussian naive bayes model on training data set 
from sklearn.naive_bayes import GaussianNB as GB
classifier_gb = GB()
classifier_gb.fit(x_train,y_train)
```

```python
# Model Accuracy on train set
train_pred_g = classifier_gb.predict(x_train)
accuracy_train_g = np.mean(train_pred_g==y_train) 
```

```python
# Model Accuracy on test set
test_pred_g = classifier_gb.predict(x_train)
accuracy_test_g = np.mean(test_pred_g==y_train)
```
