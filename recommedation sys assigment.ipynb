```python
import pandas as pd 
import numpy as np 
```

```python
book_df=pd.read_csv('book.csv',encoding='ISO-8859-1',index_col=0)
book_df
```

```python
len(book_df.userid.unique())
```

```python
len(book_df.booktitle)
```

```python
user_book_df=(book_df.pivot_table(index='userid',columns='booktitle',values='bookrating')).reset_index(drop=True)
user_book_df
```

```python
user_book_df.fillna(0,inplace=True)
user_book_df
```

```python
from sklearn.metrics import pairwise_distances
book_df_sim=1-pairwise_distances(user_book_df.values,metric='cosine')
```

```python
book_df_sim
```

```python
df_sim=pd.DataFrame(book_df_sim)
```

```python
df_sim.index=book_df.userid.unique()
df_sim.columns=book_df.userid.unique()
```

```python
df_sim.iloc[0:10,0:10]
```

```python
np.fill_diagonal(book_df_sim,0)
df_sim.iloc[0:5,0:5]
```
